<span style="color: green"><b>DEFINITION</b></span> A complex number is an ordered pair of real numbers $(a,b)$, which can be written in the form $a + bi$, where $a$ is the real part and $b$ is the imaginary part.

The symbols $z$ and $w$ are often used to denote complex numbers, such as $z = 3 + 4i$ and $w = 7 - 3i$.

Real numbers are a subset of complex numbers, where the imaginary part is zero. For example, the real number $5$ can be represented as the complex number $5 + 0i$.

Complex numbers can be added, subtracted, multiplied, and divided using the usual algebraic rules, along with the property of the imaginary unit.

The set of all complex numbers is denoted by $\mathbb{C} = \{ a $\oplus$ b i \mid a, b \in \mathbb{R} \}$.

<span style="color: orange"><b>PROPERTIES</b></span> Complex number arithmetic follows these properties for all $z, z_1, z_2, z_3 \in \mathbb{C}$ and $0,1 \in \mathbb{R}$

1. Commutative properties
    - $z_1 + z_2 = z_2 + z_1$
    - $z_1 z_2 = z_2 z_1$
2. Associative properties
    - $(z_1 + z_2) + z_3 = z_1 + (z_2 + z_3)$
    - $(z_1 z_2) z_3 = z_1 (z_2 z_3)$
3. Identities
    - There exists an additive identity $0 \in \mathbb{C}$ such that $z + 0 = z$ for all $z \in \mathbb{C}$.
    - There exists a multiplicative identity $1 \in \mathbb{C}$ such that $z \cdot 1 = z$ for all $z \in \mathbb{C}$.
4. Additive inverse
    - For each $z \in \mathbb{C}$, there exists an additive inverse $-z \in \mathbb{C}$ such that $z + (-z) = 0$.
5. Multiplicative inverse
    - For each $z \in \mathbb{C}$, $z \neq 0$, there exists a multiplicative inverse $z^{-1} \in \mathbb{C}$ such that $z z^{-1} = 1$.
6. Distributive property
    - $z_1 (z_2 + z_3) = z_1 z_2 + z_1 z_3$
    - $(z_1 + z_2) z_3 = z_1 z_3 + z_2 z_3$

For any unique complex number $z$ there exists a unique addtive inverse $-z$ such that $z + (-z) = 0$. This defines subtraction in $\mathbb{C}$ as $z_1 - z_2 = z_1 + (-z_2)$. Subtraction is the addition of the additive inverse.

The multiplicative inverse of a non-zero complex number $z$ is $z^{-1}$ such that $z z^{-1} = 1$. This defines division in $\mathbb{C}$ as $z_1 / z_2 = z_1 \cdot z_2^{-1}$ for $z_2 \neq 0$. Division is the multiplication by the multiplicative inverse.

The sets of real numbers $\mathbb{R}$ and complex numbers $\mathbb{C}$ are both fields, denoted by $\mathbb{F}$. Any property or theorem related to fields applies to both $\mathbb{R}$ and $\mathbb{C}$.

<span style="color: green"><b>DEFINITION</b></span> An element of a field $\mathbb{F}$ is called a scalar.

For any $\alpha \in \mathbb{F}$ and $m \in \mathbb{N}$, $\alpha^{m}$ is defined as the product of $m$ copies of $\alpha$, shown in (1).

$$
\alpha^{m} = \underbrace{\alpha \cdot \alpha \cdot \cdots \cdot \alpha}_{m \text{ times}}
\tag{1}
$$

This implies that $\left( \alpha^{m} \right)^{n} = \alpha^{mn}$ and that $\left( \alpha \beta \right)^{m} = \alpha^{m} \beta^{m}$ for any $\alpha, \beta \in \mathbb{F}$ and $m,n \in \mathbb{N}$.

<span style="color: green"><b>DEFINITION</b></span> If $n \in \mathbb{N}$, the order list of $n$ elements (numbers, lists, abstract objects) is called an $n$-tuple of length $n$.

An example is $\mathbb{R}^{2}$ which is the set of all ordered pairs ($2$-tuples) of real numbers, $\mathbb{R}^{2} = \{ (x,y) \mid x,y \in \mathbb{R} \}$.

Two $n$-tuples are equal if and only if their corresponding elements are the same (same elements and length).

<span style="color: green"><b>DEFINITION</b></span> $\mathbb{F}^{n}$ is the set of all $n$-tuples of elements from the field $\mathbb{F}$, where $n \in \mathbb{N}$. That is, $\mathbb{F}^{n} = \left\{ \left( x_{1} , x_{2} , \ldots , x_{n} \right) \mid x_{i} \in \mathbb{F} \text{ for } i=1,2,\ldots,n \right\}$. Here $x_{k}$ is the $k^{\text{th}}$ element of the $n$-tuple for $k = \left( 1, 2, \ldots , n \right)$.

Addition in $\mathbb{F}^{n}$ is defined as the sum of corresponding elements of two $n$-tuples. That is, for any $\mathbf{u} = (u_1, u_2, \ldots, u_n)$ and $\mathbf{v} = (v_1, v_2, \ldots, v_n)$ in $\mathbb{F}^{n}$, their sum $\mathbf{u} + \mathbf{v}$ is given in (2).

$$
\mathbf{u} + \mathbf{v} = (u_1 + v_1, u_2 + v_2, \ldots, u_n + v_n)
\tag{2}
$$

In the exercises below, we represent the $n$-tuples as column vectors. Column and row vectors are $n$-tuples, where the elements are scalars. As such, we will refer to $n$-tuples as vectors moving forward (unless otherwise specified).

Therefor, for any $\mathbf{u}, \mathbf{v} \in \mathbb{F}^{n}$, we have $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$.

The $n$-tupe of zero elements in $\mathbb{F}^{n}$ is called the zero vector and is denoted by $\mathbf{0} = (0,0,\ldots,0)$, where $0$ is the additive identity in the field $\mathbb{F}$. For any $\mathbf{u} \in \mathbb{F}^{n}$, we have $\mathbf{u} + \mathbf{0} = \mathbf{u}$.

Scalar multiplication in $\mathbb{F}^{n}$ is defined as the product of a scalar from the field $\mathbb{F}$ and each element of an $n$-tuple. That is, for any $\alpha \in \mathbb{F}$ and $\mathbf{u} = (u_1, u_2, \ldots, u_n) \in \mathbb{F}^{n}$, their product $\alpha \mathbf{u}$ is given in (3).

$$
\alpha \mathbf{u} = (\alpha u_1, \alpha u_2, \ldots, \alpha u_n)
\tag{3}
$$

## VECTOR SPACES

<span style="color: green"><b>DEFINITION</b></span> A vector space over a field $\mathbb{F}$ is a non-empty set $V$ together with two operations: vector addition $\oplus$ and scalar multiplication (juxtaposition). These operations must satisfy the following axioms for all $\mathbf{u}, \mathbf{v}, \mathbf{w} \in V$ and $\alpha, \beta \in \mathbb{F}$.

1. Commutativity of addition: $\mathbf{u} \oplus \mathbf{v} = \mathbf{v} \oplus \mathbf{u}$
2. Associativity of addition: $(\mathbf{u} \oplus \mathbf{v}) \oplus \mathbf{w} = \mathbf{u} \oplus (\mathbf{v} \oplus \mathbf{w})$
3. Existence of additive identity: There exists an element $\mathbf{e} \in V$ such that $\mathbf{u} \oplus \mathbf{e} = \mathbf{u}$
4. Existence of additive inverse: For each $\mathbf{u} \in V$, there exists an element $-\mathbf{u} \in V$ such that $\mathbf{u} \oplus (-\mathbf{u}) = \mathbf{e}$
5. Compatibility of scalar multiplication with field multiplication: $\alpha (\beta \mathbf{u}) = (\alpha \beta) \mathbf{u}$
6. Identity element of scalar multiplication: $1 \mathbf{u} = \mathbf{u}$, where $1$ is the multiplicative identity in $\mathbb{F}$
7. Distributivity of scalar multiplication with respect to vector addition: $\alpha (\mathbf{u} \oplus \mathbf{v}) = \alpha \mathbf{u} \oplus \alpha \mathbf{v}$
8. Distributivity of scalar multiplication with respect to field addition: $(\alpha \oplus \beta) \mathbf{u} = \alpha \mathbf{u} \oplus \beta \mathbf{u}$

The additive and multiplicative identity depend on the field $\mathbb{F}$ over which the vector space is defined and on the set $V$ itself.

A vector space over the field $\mathbb{R}$ is called a real vector space and a vector space over the field $\mathbb{C}$ is called a complex vector space.

An example of a vector space is $\mathbb{F}^{\infty}$, the set of all sequences of elements from the field $\mathbb{F}$, with the operations of vector addition and scalar multiplication defined in (4), for all $x_{i}, y_{i}, \alpha \in \mathbb{F}$.

$$
\begin{align*}
\left( x_{1} , x_{2}, x_{3}, \ldots \right) + \left( y_{1} , y_{2}, y_{3}, \ldots \right) &= \left( x_{1} + y_{1} , x_{2} + y_{2}, x_{3} + y_{3}, \ldots \right) &\text{(addition)}\\ \\
\alpha \left( x_{1} , x_{2}, x_{3}, \ldots \right) &= \left( \alpha x_{1} , \alpha x_{2}, \alpha x_{3}, \ldots \right) &\text{(scalar multiplication)}
\end{align*}
\tag{4}
$$

$\mathbb{F}^{n}$ is a special case of $\mathbb{F}^{\infty}$ where all elements after the $n^{\text{th}}$ element are zero.

If $S$ is a set, then $\mathbb{F}^{S}$ is the set of all functions from $S$ to the field $\mathbb{F}$. That is, $\mathbb{F}^{S} = \{ \theta \mid \theta : S \to \mathbb{F} \}$. The operations of vector addition and scalar multiplication in $\mathbb{F}^{S}$ are defined as follows for all $f, g \in \mathbb{F}^{S}$, $\alpha \in \mathbb{F}$, and $x \in S$, shown in (5).

$$
\begin{align*}
\left( f + g \right)(x) & = f(x) + g(x) &\text{(addition)}\\ \\
\left( \alpha f \right)(x) & = \alpha f(x) &\text{(scalar multiplication)}
\end{align*}
\tag{5}
$$

<span style="color: darkred"><b>THEOREM</b></span> The additiove identiy in a vector space $V$ is unique.

__Proof__

Let $\mathbf{e}$ and $\mathbf{e}^{*}$ be any two arbitrary additive identities in the vector space $V$ such that $\mathbf{u} + \left( - \mathbf{u} \right) = \mathbf{e}$ and $\mathbf{u} + \left( - \mathbf{u} \right) = \mathbf{e}^{*}$, for any $\mathbf{u} \in V$. The derivation in (6) follows.

$$
\begin{align*}
&\mathbf{u} + \left( -\mathbf{u} \right) = \mathbf{e}^{*} \\ \\
&\mathbf{e} = \mathbf{e}^{*} &\text{(subsitute for } \mathbf{u} + \left( - \mathbf{u} \right) \text{ by definition)}
\end{align*}
\tag{6}
$$

$\square$

<span style="color: darkred"><b>THEOREM:</b></span> Every element $\mathbf{u}$ in a vector space $V$ has a unique additive inverse $-\mathbf{u}$.

__Proof__

Let $-\mathbf{u}$ and $-\mathbf{u}^{*}$ be any two arbitrary additive inverses of the element $\mathbf{u}$ in the vector space $V$ such that $\mathbf{u} + \left( - \mathbf{u} \right) = \mathbf{e}$ and $\mathbf{u} + \left( - \mathbf{u}^{*} \right) = \mathbf{e}$, where $\mathbf{e}$ is the additive identity in $V$. The derivation in (7) follows.

$$
\begin{align*}
&-\mathbf{u} = -\mathbf{u} + \mathbf{e} &\text{(by definition of a vector space)} \\ \\
&-\mathbf{u} = -\mathbf{u} + \left( \mathbf{u} + \left( - \mathbf{u}^{*} \right) \right) &\text{(subsitute for } \mathbf{e} \text{ by definition)} \\ \\
&-\mathbf{u} = \left( -\mathbf{u} + \mathbf{u} \right) + \left( - \mathbf{u}^{*} \right) &\text{(by associativity of addition)} \\ \\
&-\mathbf{u = \mathbf{e} + \left( - \mathbf{u}^{*} \right)} &\text{(by definition of an additive inverse)} \\ \\
&-\mathbf{u} = - \mathbf{u}^{*} &\text{(by definition of a vector space)}
\end{align*}
\tag{7}
$$

$\square$

<span style="color: darkred"><b>THEOREM:</b></span> $0 \mathbf{u} = \mathbf{0}$ for any $\mathbf{u} \in V$ and $0 \in \mathbb{F}$, and where $\mathbf{0}$ is the zero vector (additive identity) in $V$.

__Proof__

The proof follows in (8).

$$
\begin{align*}
&0 \mathbf{u} = \left( 0 + 0 \right) \mathbf{u} &\text{(by definition of a field)} \\ \\
&0 \mathbf{u} = 0 \mathbf{u} + 0 \mathbf{u} &\text{(by distributivity of scalar multiplication)} \\ \\
&0 \mathbf{u} + \left( - 0 \mathbf{u} \right) = \left( 0 \mathbf{u} + 0 \mathbf{u} \right) + \left( - 0 \mathbf{u} \right) &\text{(by definition of a vector space)} \\ \\
&\mathbf{0} = 0 \mathbf{u} + \left( 0 \mathbf{u} + \left( - 0 \mathbf{u} \right) \right) &\text{(by definition of an additive inverse)} \\ \\
&\mathbf{0} = 0 \mathbf{u} + \mathbf{0} &\text{(by definition of an additive inverse)} \\ \\
&\mathbf{0} = 0 \mathbf{u} &\text{(by definition of a vector space)}
\end{align*}
\tag{8}
$$

$\square$

<span style="color: darkred"><b>THEOREM:</b></span> For any $\mathbf{u} \in V$, $(-1) \mathbf{u} = -\mathbf{u}$, where $-1$ is the additive inverse of $1$ in the field $\mathbb{F}$.

__Proof__

The proof follows in (9).

$$
\begin{align*}
&\mathbf{u} + (-1) \mathbf{u} = 1 \mathbf{u} + (-1) \mathbf{u} &\text{(by definition of a vector space)} \\ \\
&\mathbf{u} + (-1) \mathbf{u} = \left( 1 + (-1) \right) \mathbf{u} &\text{(by distributivity of scalar multiplication)} \\ \\
&\mathbf{u} + (-1) \mathbf{u} = 0 \mathbf{u} &\text{(by definition of a field)} \\ \\
&\mathbf{u} + (-1) \mathbf{u} = \mathbf{0} &\text{(by Theorem above)} \\ \\
&(-1) \mathbf{u} = -\mathbf{u} &\text{(by definition of an additive inverse)}
\end{align*}
\tag{9}
$$

$\square$

<span style="color: darkred"><b>THEOREM:</b></span> $-(- \mathbf{u}) = \mathbf{u}$ for all $\mathbf{u} \in V$.

__Proof__

The proof follows in (10).

$$
\begin{align*}
&- (- \mathbf{u}) = -1 (- \mathbf{u}) &\text{(by Theorem above)} \\ \\
&- (- \mathbf{u}) = (-1)(-1) \mathbf{u} &\text{(by Theorem above)} \\ \\
&- (- \mathbf{u}) = 1 \mathbf{u} &\text{(by multiplicative property of real numbers)} \\ \\
&- (- \mathbf{u}) = \mathbf{u} &\text{(by definition of a vector space)}
\end{align*}
\tag{10}
$$

<span style="color: darkred"><b>THEOREM:</b></span> $a \mathbf{0} = \mathbf{0}$ for any $a \in \mathbb{F}$, where $\mathbf{0}$ is the zero vector (additive identity) in $V$.

__Proof__

The proof follows in (11) for any $a \in \mathbb{F}, a \ne 0$. (We have already proven the case for $a = 0$.)

$$\begin{align*}
&a \mathbf{0} = a \left( 0 \mathbf{u} \right) &\text{(by Theorem above)} \\ \\
&a \mathbf{0} = \left( a \cdot 0 \right) \mathbf{u} &\text{(by associativity of scalar multiplication)} \\ \\
&a \mathbf{0} = 0 \mathbf{u} &\text{(by definition of a field)} \\ \\
&a \mathbf{0} = \mathbf{0} &\text{(by Theorem above)}
\end{align*}
\tag{11}
$$

$\square$

<span style="color: darkred"><b>COROLLARY:</b></span> If $a \mathbf{u} = \mathbf{0}$ for some $a \in \mathbb{F}$ and $\mathbf{u} \in V$, then either $a = 0$ or $\mathbf{u} = \mathbf{0}$.

__Proof__

The proof follows from previous theorems that shows that $a \mathbf{0} = \mathbf{0}$ for any $a \in \mathbb{F}, a \ne 0$ and that $0 \mathbf{u} = \mathbf{0}$ for any $\mathbf{u} \in V$.

$\square$

<span style="color: royalblue"><b>EXERCISE:</b></span> Why is the empty set not a vector space?

The empty set does not contain an additive identity, which is required by the definition of a vector space. The set that only contains the zero vector is the smallest vector space possible.
